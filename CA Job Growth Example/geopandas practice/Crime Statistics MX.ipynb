```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt

%matplotlib inline

plt.rcParams['figure.figsize'] = 10,10
```

```python
df = pd.read_csv('nm-fuero-comun-municipios.csv')
```

```python
df.head()
```

```python
df['IDUNICO'] = df['state_code']*1000 + df['mun_code']
```

```python
df[['state','municipio','IDUNICO']]
```

```python
municipios[municipios['CVE_ENT'] == 1]
```

```python
df['date'] = pd.to_datetime(df['date'], format = '%Y-%m', )
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
gdf_merged = municipios.merge(df, left_on='IDUNICO', right_on='IDUNICO', how='inner')
```

```python
gdf_merged.head()
```

```python
feminicidios =  gdf_merged[gdf_merged['tipo'] == 'FEMINICIDIO'].copy()
homicidios = gdf_merged[gdf_merged['tipo'] == 'HOMICIDIO'].copy()
```

```python
feminicidios.shape
```

```python
homicidios.shape
```

```python
homicidios[homicidios['year'] == 2016].plot(column = 'count')
```

```python
homicidios17 = homicidios[homicidios['year']==2017].copy()
```

```python
homicidios17.shape
```

```python
dolosos17 = homicidios17[homicidios17['subtipo'] == 'HOMICIDIO DOLOSO'].copy()
```

```python
dolosos17.shape
```

```python
armas = dolosos17[dolosos17['modalidad'] == 'CON ARMA DE FUEGO'].copy()
```

```python
armas[armas['month'] == 4].plot(column = 'count', legend = True, figsize = (20,20), cmap = 'Reds')
```

```python
import json
import altair as alt
```

```python
armas['date'] = armas['date'].astype(object)
```

```python
armas.drop('date', axis = 1, inplace=True)
```

```python
# dump as json
json_gdf = armas[armas['month'] == 4].to_json()
# load json to select variable `features` and dump as json
json_features = json.dumps(json.loads(json_gdf)['features'])
```

```python
data_geo = alt.Data(values=json_features)

# plot map, where variables ares nested within `properties`, 
alt.Chart(data_geo).mark_geoshape(
    fill='lightgray',
    stroke='white',
    color = 'red',
).properties(
    projection={'type': 'albers'},
    width=700,
    height=400,
    
).encode(
    color=alt.Color('properties.count:Q', legend = alt.Legend(orient='left', title = 'Homicidios Dolosos (Abril 2017)', scale = {'scheme': 'reds'})),
)
```
